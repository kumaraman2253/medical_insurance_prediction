```python
import pandas as pd
```

```python
df=pd.read_csv('insurance.csv')
```

```python
df.head(10)
```

```python
#to change char to float data type using label encoder
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex=LabelEncoder()
le_region=LabelEncoder()
le_smoker=LabelEncoder()
```

```python
df['sex_new']=le_sex.fit_transform(df['sex'])
df['region_new']=le_region.fit_transform(df['region'])
df['smoker_new']=le_smoker.fit_transform(df['smoker'])
#for sex- female=0, male=1
#for region- southwest=3, southeast=2, northwest=1, northeast=0
#for smoker- yes=1, no=0
```

```python
df.head(10)
```

```python
inputs=df[['age','sex_new','bmi','children','smoker_new','region_new']]
inputs.head(10)
```

```python
inputs.isna()
```

```python
target=df[['charges']]
target.head(10)
```

```python
target.isna()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(inputs,target,test_size=0.2,random_state=500)
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
#creation of our Linear regression model
```

```python
reg.score(x_test,y_test)
```
